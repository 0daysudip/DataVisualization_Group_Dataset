```python
import pandas as pd

AQI     = r"C:\Users\asus\Desktop\New_Dataset\AQI\AQI_Pokhara_Ward7_Cleaned.csv"
WEATHER = r"C:\Users\asus\Desktop\New_Dataset\WEATHER\Weather_Pokhara_Cleaned.csv"
OUTPUT  = r"C:\Users\asus\Desktop\New_Dataset\COMBINED_POKHARA.csv"

aqi     = pd.read_csv(AQI)
weather = pd.read_csv(WEATHER)

aqi["date"] = pd.to_datetime(aqi["datetime_utc"], utc=True).dt.strftime("%Y-%m-%d")

df = aqi.merge(
    weather[["date", "rainfall (mm)", "humidity (%)", "windspeed (km/h)"]],
    on="date", how="left"
)

df = df[df["date"] <= "2026-06-30"]

df = df[[
    "sensor_id",
    "datetime_utc",
    "datetime_local",
    "Parameter(pm1)",
    "Parameter(pm2.5)",
    "temperature",
    "relativehumidity",
    "rainfall (mm)",
    "humidity (%)",
    "windspeed (km/h)",
    "city",
    "latitude",
    "longitude",
    "date",
]]

df = df.dropna()
df = df.sort_values("datetime_utc").reset_index(drop=True)
df.to_csv(OUTPUT, index=False)

print(f"Saved : {OUTPUT}")
print(f"Rows  : {len(df):,}")
print(f"Cols  : {df.columns.tolist()}")
print(f"From  : {df['date'].min()} to {df['date'].max()}")
print(df.head(3).to_string())
```
