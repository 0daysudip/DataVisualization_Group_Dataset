```python
import pandas as pd

KATHMANDU = r"C:\Users\asus\Desktop\New_Dataset\COMBINED\COMBINED_KATHMANDU.csv"
POKHARA   = r"C:\Users\asus\Desktop\New_Dataset\COMBINED\COMBINED_POKHARA.csv"
SURKHET   = r"C:\Users\asus\Desktop\New_Dataset\COMBINED\COMBINED_SURKHET.csv"
OUTPUT    = r"C:\Users\asus\Desktop\New_Dataset\COMBINED\FINAL_CLEANED.csv"

kathmandu = pd.read_csv(KATHMANDU)
pokhara   = pd.read_csv(POKHARA)
surkhet   = pd.read_csv(SURKHET)

print(f"Kathmandu rows : {len(kathmandu):,}")
print(f"Pokhara rows   : {len(pokhara):,}")
print(f"Surkhet rows   : {len(surkhet):,}")

df = pd.concat([kathmandu, pokhara, surkhet], ignore_index=True)

df = df.sort_values(["city", "datetime_utc"]).reset_index(drop=True)

df.to_csv(OUTPUT, index=False)

print(f"\nSaved  : {OUTPUT}")
print(f"Rows   : {len(df):,}")
print(f"Cols   : {df.columns.tolist()}")
print(f"Cities : {df['city'].unique().tolist()}")
print(f"From   : {df['date'].min()} to {df['date'].max()}")
print(f"\nRows per city:")
print(df.groupby("city").size().to_string()) 
```
