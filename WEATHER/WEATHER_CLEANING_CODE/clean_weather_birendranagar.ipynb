```python
import pandas as pd

INPUT  = r"C:\Users\asus\Desktop\New_Dataset\WEATHER\Weather_Birendranagar.csv"
OUTPUT = r"C:\Users\asus\Desktop\New_Dataset\WEATHER\Weather_Birendranagar_Cleaned.csv"

df = pd.read_csv(INPUT)

print("Raw rows     :", len(df))
print("Raw columns  :", df.columns.tolist())
print()
print(df.head(3).to_string())

df["sensor_id"] = 14476518

df["date"] = pd.to_datetime(df["date"]).dt.strftime("%Y-%m-%d")

df = df.rename(columns={
    "rainfall_mm"   : "rainfall (mm)",
    "humidity_pct"  : "humidity (%)",
    "windspeed_kmh" : "windspeed (km/h)",
})

df = df.dropna()

df = df[df["rainfall (mm)"].between(0, 400)]
df = df[df["humidity (%)"].between(0, 100)]
df = df[df["windspeed (km/h)"].between(0, 150)]

df = df[["sensor_id", "date", "rainfall (mm)", "humidity (%)", "windspeed (km/h)"]]

df = df.sort_values("date").reset_index(drop=True)

df.to_csv(OUTPUT, index=False)

print(f"\nCleaned rows : {len(df):,}")
print(f"Columns      : {df.columns.tolist()}")
print(f"Saved to     : {OUTPUT}")
print()
print(df.head(10).to_string())
```
